Euclidean distance function to calculate the distaance between images 

```python
def euclidean_distance(image, test):
    distance  = np.sqrt ( np.sum( (image-test) ** 2) )
    return distance
```

Function that calculates the distances between images and test

```python
def Image_distances(image1,image2,image3,test):
  
  distance1 = euclidean_distance(image1,test)
  distance2 = euclidean_distance(image2,test)
  distance3 = euclidean_distance(image3,test)
  average_distance = sum(distance1 ,distance2 ,distance3)
  average_distance  = average_distance/3

  return average_distance
```
